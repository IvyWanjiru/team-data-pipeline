```python
# Student B(Ivy)
import numpy as np
import pandas as pd

# 1. Create a NumPy array tracking processing times (in seconds) for 5 data batches
processing_times = np.array([45.2, 50.1, 42.8, 61.3, 48.5])

# 2. Create a Pandas DataFrame tracking team task completion metrics
pipeline_data = {
    'Task_ID': [1, 2, 3, 4, 5],
    'Team_Member': ['Student A', 'Student B', 'Student C', 'Student A', 'Student B'],
    'Lines_of_Code': [120, 85, 150, 95, 110],
    'Errors_Found': [3, 1, 0, 4, 2]
}

df = pd.DataFrame(pipeline_data)

# Display the created data structures
print("--- NumPy Array (Processing Times) ---")
print(processing_times)
print("\n--- Pandas DataFrame (Pipeline Tasks) ---")
print(df)
```

```python
# Student B(Ryan)-  Descriptive statistics on the pipeline data

# Average processing time from the NumPy array
avg_processing_time = processing_times.mean()
print("Average processing time (seconds):", avg_processing_time)

# Descriptive statistics on the DataFrame's numeric columns
print("\n Descriptive Statistics (Lines_of_Code, Errors_Found) ")
print(df.describe())

# Average lines of code and errors found
avg_loc = df['Lines_of_Code'].mean()
avg_errors = df['Errors_Found'].mean()
print("\nAverage Lines of Code:", avg_loc)
print("Average Errors Found:", avg_errors)
```
